```python
# Eliminamos columnas innecesarias para nuestro experimiento
#Some basic Feature engineering
df = df.drop(['Site','Unnamed: 2', 'Unnamed: 3', 'Contexto', 'suma'], axis=1)
#Comprobamos que no hay valores faltantes
# Checking missing values
df.isnull().any().any()
```

```python
df.info()
```

```python
df['rand'] = np.random.normal(loc=30, scale=1.5, size=1816)

# Limitar los valores entre 0.70 y 59.0
df['rand'] = df['rand'].clip(lower=0.70, upper=59.0)
```

```python
df.info()
```

```python
# Distribución en el target_Group
# More Exploratory Analysis
target = df['target_PDLC'].value_counts()
target
```

```python
X = df.drop(['target_PDLC'], axis=1)
y = df['target_PDLC']
```

```python
# oversampling strategy
smote = SMOTE()
X_smote, y_smote = smote.fit_resample(X, y)

y_smote.value_counts().plot(kind='bar')
#plt.savefig('SMOTE_distribution.png',dpi=200, bbox_inches='tight')
y_smote.value_counts()
```

```python
# Creating final resampled dataset
resampled_df = X_smote
resampled_df['target'] = y_smote
```

```python
resampled_df['target'].value_counts()
```

```python
# Validation splitting
data = resampled_df.sample(frac=0.90, random_state=786)
fvsAliste = resampled_df.drop(data.index)
data.reset_index(inplace=True, drop=True)
fvsAliste.reset_index(inplace=True, drop=True)
print('Data for further procedures: ' + str(data.shape))
print('Data for final validation: ' + str(fvsAliste.shape))
#fvsAliste.to_csv('./TABLES/final_validation_set_aliste.csv')
```
