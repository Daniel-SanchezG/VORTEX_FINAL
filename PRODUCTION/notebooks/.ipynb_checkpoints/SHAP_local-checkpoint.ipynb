```python
#Some basic Feature engineering
df = df.drop(['Site', 'suma'], axis=1)
#Comprobamos que no hay valores faltantes
# Checking missing values
df.isnull().any().any()
```

```python
# Distribución en el target_Group
# More Exploratory Analysis
target = df['target_PDLC'].value_counts()
target
```

```python
X = df.drop(['target_PDLC'], axis=1)
y = df['target_PDLC']
```

```python
# oversampling strategy
smote = SMOTE()
X_smote, y_smote = smote.fit_resample(X, y)

y_smote.value_counts().plot(kind='bar')
#plt.savefig('SMOTE_distribution.png',dpi=200, bbox_inches='tight')
y_smote.value_counts()
```

```python
# Creating final resampled dataset
data = X_smote
data['target'] = y_smote
```

```python
data['target'].value_counts()
```

```python
# Experiment setup 

experiment = setup(data=data, target= 'target',train_size=0.8, session_id=123)
```

```python
remove_metric('MCC')
remove_metric('Kappa')
remove_metric('AUC')
#'AUC', , 'MCC'
```

```python
rflocal = create_model('rf',n_estimators=200, min_samples_leaf=5, class_weight="balanced", criterion='entropy')
```

### Local Shap Value Analysis

```python
import shap
td = get_config(variable="X_train")
explainer = shap.Explainer(rflocal)
shap_values = explainer(td)
```

```python
from shap.plots import beeswarm
beeswarm(shap_values[:,:,1],max_display=20, color=plt.get_cmap("viridis")) #save this plot
```

```python

shap.plots.beeswarm(shap_values[:,:,1], order=shap_values.abs.max(0))

```

```python

tuned_explainer = shap.Explainer(tuned_local_model)
tuned_shap = tuned_explainer(td)

beeswarm(tuned_shap[:,:,1],max_display=20,  color=plt.get_cmap("viridis")) #save this plot
```

```python
# Solución para el error con shap.plots.beeswarm(shap_values.abs)
# El problema es que shap_values tiene 3 dimensiones y .abs mantiene esa estructura

# Opción 1: Usar solo los valores absolutos para una clase específica (en este caso, clase 1)
shap.plots.beeswarm(shap_values[:,:,1].abs, color="shap_red",max_display=20)

```

```python
# Modificación del método 2 para crear un objeto Explanation adecuado
from shap import Explanation

# Calculamos los valores absolutos
abs_values = np.abs(shap_values.values)
total_magnitude = abs_values[:,:,0] + abs_values[:,:,1]

# Creamos un objeto Explanation con los valores de magnitud total
explanation = Explanation(
    values=total_magnitude,  # Valores de magnitud total
    base_values=np.mean(total_magnitude),  # Valor base aproximado
    data=td.values,  # Datos originales
    feature_names=td.columns  # Nombres de características
)

# Ahora usamos el objeto Explanation para el gráfico
shap.plots.beeswarm(explanation, color="shap_red")
```

```python
# Opción 3: Convertir a un ExplanationDashboard más simple
from shap import Explanation
# Crear un nuevo objeto Explanation con los valores absolutos
explanation = Explanation(
    values=np.abs(shap_values.values[:,:,1]),  # Usar solo clase 1
    base_values=np.abs(shap_values.base_values[:,1]),  # Base values para clase 1
    data=td.values,  # Datos originales
    feature_names=td.columns  # Nombres de características
)
shap.plots.beeswarm(explanation, color="shap_red", max_display=20)
```
