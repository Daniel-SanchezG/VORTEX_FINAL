#### Filtering features 

We are going to perform a feature selection task based on the previous iterations.

```python
#Subsets to train specific models

features_pool ={
'trainning_features_Destilled': ['Ca','S', 'K', 'Ti', 'V', 'Cr', 'Cu', 
                      'Zn', 'As', 'Se', 'Sr', 
                      'Mo','Ba', 
                      'Ta', 'Site'],
'trainning_features_french': ['Al','Si','P','S','Cl','K','Ca',
                      'Ti','V','Cr','Fe','Co','Cu','Zn',
                      'As','Se','Rb','Sr','Zr','Mo','Ba',
                      'Ta','Site'],

'trainning_features_Quiruelas': ['Al', 'Si', 'P','K','Ca', 
                      'Ti', 'V', 'Cr', 'Mn','Fe','Co',
                      'Ga','Zn', 'As', 'Rb', 
                      'Sr', 'Zr', 'Site'],

'trainning_features_VdH': ['Al', 'Si', 'P', 'S', 'Cl', 'K',
                      'Ti', 'V', 'Cr', 'Mn','Fe', 'Co', 
                      'Ni','Cu','Zn', 'As', 'Rb', 
                      'Sr', 'Zr', 'Site']
}
```

```python
# Import specific libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
import pandas as pd
import pickle
import os
import matplotlib.pyplot as plt

# Create necessary directories
output_dirs = {
    'models': 'models',
    'tables': 'tables',
    'plots': 'plots'
}

# Ensure that all directories exist.
for directory in output_dirs.values():
    if not os.path.exists(directory):
        os.makedirs(directory)

# Dictionary for storing trained and calibrated models
trained_models = {}
calibrated_models = {}
evaluation_reports = {}

# Initialise SMOTE
smote = SMOTE(random_state=42)

# Iterate over each configuration in features_pool
for config_name, features in features_pool.items():
    print(f"\n{'='*50}")
    print(f"Processing configuration: {config_name}")
    print(f"{'='*50}")
    
    # Extract the subset of characteristics for this model
    data_subset = data_trainning[features]
    
    # Prepare the validation set 
    fvs_subset = fvs[features]
    
    # Splitting data for training
    X = data_subset.drop(['Site'], axis=1)
    y = data_subset['Site']
    
    # Splitting data for validación
    X_val = fvs_subset.drop(['Site'], axis=1)
    y_val = fvs_subset['Site']
    
    print(f"training data shape: {X.shape}, class distribution: {y.value_counts()}")
    print(f"Validation data shape: {X_val.shape}, class distribution: {y_val.value_counts()}")
    
    # Aplying SMOTE 
    X_balanced, y_balanced = smote.fit_resample(X, y)

    print(f"Balanced data shape: {X_balanced.shape}, class distribution: {y_balanced.value_counts()}")
    
    # Calibration split
    X_train, X_calib, y_train, y_calib = train_test_split(
        X_balanced, y_balanced,
        test_size=0.2,
        random_state=42,
        stratify=y_balanced
    )
    
    # Create a Random Forest Classifier
    model_name = config_name.replace('trainning_features_', 'rf_')
    print(f"\nTraining model {model_name}...")
    
    rf = RandomForestClassifier(
        n_estimators=200,
        min_samples_leaf=5,
        class_weight="balanced",
        criterion='entropy',
        random_state=42
    )
    
    # Training the model
    rf.fit(X_train, y_train)
    
    # Save the base model in the dictionary
    trained_models[model_name] = rf
    
    # Calibrate the model
    print(f"Calibrating model {model_name}...")
    calibrator = CalibratedClassifierCV(rf, cv='prefit', method='sigmoid')
    calibrator.fit(X_calib, y_calib)
    
    # Save the calibrated model
    calibrated_models[model_name] = calibrator
    
    # Evaluating model on vlaidation data
    print(f"Evaluating model {model_name} on validation set...")
    y_pred = calibrator.predict(X_val)
    
    # Create prediction DataFrame
    fvs_predictions = fvs_subset.copy()
    fvs_predictions['predictions'] = y_pred
    
    # Generate and save classificaiton report
    y_true = fvs_predictions['Site']
    class_rep = pd.DataFrame(classification_report(y_true, y_pred, output_dict=True)).T
    
    # Save report on the dicitonary
    evaluation_reports[model_name] = class_rep
    
    # Save report as CSV
    report_path = os.path.join(output_dirs['tables'], f"{model_name}_classification_report.csv")
    class_rep.to_csv(report_path)
    print(f"Classification report saved in: {report_path}")
    
    # Generate and save confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    disp = ConfusionMatrixDisplay(
        confusion_matrix=cm,
        display_labels=['Can_tintorer', 'Terena', 'Aliste']
    )
    disp.plot(cmap='Greens')
    plt.title(f"Confusion matrix for {model_name}")
    
    # Save confusion matrix
    confusion_matrix_path = os.path.join(output_dirs['plots'], f"{model_name}_confusion_matrix.png")
    plt.savefig(confusion_matrix_path, bbox_inches='tight')
    plt.close()
    print(f"Confusion matrix saved in: {confusion_matrix_path}")
    
    # Save calibrates model for re-use
    model_file = os.path.join(output_dirs['models'], f"{model_name}_calibrated.pkl")
    with open(model_file, 'wb') as file:
        pickle.dump(calibrator, file)
    
    print(f"Calibrated model saved in: {model_file}")

# print performance summary
print("\n\n" + "="*80)
print("MODEL PERFORMANCE SUMMARY")
print("="*80)

for model_name, report in evaluation_reports.items():
    print(f"\nModel: {model_name}")
    print(f"Precisión: {report.loc['accuracy']['f1-score']:.4f}")
    print(f"F1-score (macro avg): {report.loc['macro avg']['f1-score']:.4f}")
    print(f"F1-score (weighted avg): {report.loc['weighted avg']['f1-score']:.4f}")
    print("-"*50)

print("\n¡ProcesS completed!")
```

### Model Calibration:
This function calibrates the probability of a given estimator using isotonic or logistic regression. The output of this function is a score grid with CV scores by fold.
By default function uses Logistic regression model to transform probabilities.
We chose this method to take into account the following library documentation [warning](https://pycaret.readthedocs.io/en/latest/api/classification.html#pycaret.classification.calibrate_model) and avoid overfitting.
